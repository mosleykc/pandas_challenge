```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Print school_data_complete to look at the merged data - first five rows

school_data_complete_df.head()
```

```python
# Calculate all district data requests

# Calculate the total number of unique schools
number_of_schools = len(school_data_complete_df["school_name"].unique())

# Calculate the total number of students - use student ID because there are multiple students with same name and 
# student ID is unique 
number_of_students = student_data_df["Student ID"].count()

# Calculate the total budget for the district
total_dist_budget = school_data_df["budget"].sum()

# Calculate the average math score
avg_dist_math_score = school_data_complete_df["math_score"].mean()

# Calculate the average reading score
avg_dist_reading_score = school_data_complete_df["reading_score"].mean()

#Calculate the overall passing rate (overall average score), first calculate passing score for math and reading
passing_rate_dist_math = student_data_df.loc[student_data_df["math_score"] >= 70]["math_score"].count()
passing_rate_dist_reading = student_data_df.loc[student_data_df["reading_score"] >= 70]["reading_score"].count()

# Calculate the percentage of students with a passing math score (70 or greater)
percent_students_dist_pass_math = (passing_rate_dist_math/number_of_students)*100

# Calculate the percentage of students with a passing reading score (70 or greater)
percent_students_dist_pass_reading = (passing_rate_dist_reading/number_of_students)*100

overall_passing_score = (percent_students_dist_pass_math + percent_students_dist_pass_reading)/2
```

```python
# District Summary DataFrame Unformatted
dist_summary_unformatted_df = pd.DataFrame({
    "Total Schools" : [number_of_schools],
    "Total Students" : [number_of_students],
    "Total Budget" : [total_dist_budget],
    "Avg Math Score" : [avg_dist_math_score],
    "Avg Reading Score" : [avg_dist_reading_score], 
    "% Passing Math" : [percent_students_dist_pass_math],
    "% Passing Reading" : [percent_students_dist_pass_reading],
    "% Overall Passing Rate" : [overall_passing_score] 
})

dist_summary_unformatted_df
```

```python
# Add formatting for client presentation

Formatted_District_Summary = dist_summary_unformatted_df.style.format({
    "Total Schools":"{:,}",
    "Total Students":"{:,}",
    "Total Budget":"${:,.2f}",
    "Average Math":"{:.2f}", 
    "Average Reading":"{:.2f}",
    "% Passing Math":"{:.2f}%", 
    "% Passing Reading":"{:.2f}%", 
    "% Overall Passing Rate":"{:.2f}%"
})

Formatted_District_Summary
```

```python
# Breaking data requests in individual cells to try to correct issue causing runtime error. Note: # Total Students, 
# School Type, Total Students by School, Total School Budget, Per Student Budget can be calculated in the 
# school summary without creating it's own variable.

# List School Name - assign to new var
school_list = school_data_df["school_name"]
school_list.head(20)
```

```python
# Define school type by school and assign new var
school_type = school_data_complete_df.groupby("school_name")["type"].unique()
school_type = school_type.str[0]
school_type.head(15)
```

```python
# Define number of students per school and assign to new var. This could also by done by using the "size" value in school_data_pd 
# but feel counting indiv students by school could be more accurate.
students_per_school = student_data_df.groupby('school_name')['student_name'].count()
students_per_school.head(15)
```

```python
# define Total School Budget and assign to new var
school_budget = school_data_df.set_index('school_name')['budget']
school_budget.head()
```

```python
# define Per Student Budget
per_student_budget = school_data_df.set_index('school_name')['budget']/school_data_df.set_index('school_name')['size']

#per_student_budget = school_data_df["budget"] / school_data_df["size"]
per_student_budget.head(15)
```

```python
# group school_complete_data by School Name to reduce having to use .groupby
school_group_by_schoolname = school_data_complete_df.groupby(["school_name"])

# define Average Math Score and assign to var
avg_math_score = round(school_group_by_schoolname["math_score"].mean())

avg_math_score
```

```python
# define Average reaing Score and assign to var
avg_reading_score = round(school_group_by_schoolname["reading_score"].mean())

avg_reading_score
```

```python
#create DF that calculates and combines passing for reading and math
passing_df = school_data_complete_df.loc[(school_data_complete_df['math_score'] >=70) 
                                      & (school_data_complete_df['reading_score'] >=70)]
pass_math_df = school_data_complete_df.loc[(school_data_complete_df['math_score'] >=70)]
pass_reading_df = school_data_complete_df.loc[(school_data_complete_df['reading_score'] >=70)]
pass_reading_df.head()
```

```python
#Define % passing math
percent_pass_math = round((pass_math_df.groupby('school_name')['math_score'].count()/students_per_school)*100,1)

percent_pass_math
```

```python
#Define % passing reading
percent_pass_reading = round((pass_reading_df.groupby('school_name')['reading_score'].count()/students_per_school)*100,1)

percent_pass_reading
```

```python
# define Overall Passing Rate (Average of the above two)
overall_pass_rate = round((percent_pass_math + percent_pass_reading)/2, 2)
overall_pass_rate
```

```python
# Create unformatted dataframe
school_summary_unformatted_df = pd.DataFrame({
    #"School Name" : school_list,
    "School Type" : school_type,
    "Total Students" : students_per_school,
    "Total School Budget" : school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score" : avg_math_score,
    "Average Reading Score" : avg_reading_score,
    "% Passing Math" : percent_pass_math,
    "% Passing Reading" : percent_pass_reading,
    "Overall Passing Rate" : overall_pass_rate
        
    
})

# 
school_summary_data_df = school_summary_unformatted_df[[
    "School Type", "Total Students", "Total School Budget", "Per Student Budget", "Average Math Score", \
    "Average Reading Score", "% Passing Math", "% Passing Reading", "Overall Passing Rate"]]


school_summary_data_df.head(15)


    
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a table that lists the average MATH Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

grade_9 = student_data_df[(student_data_df["grade"] == "9th")]
grade_10 = student_data_df[(student_data_df["grade"] == "10th")] 
grade_11 = student_data_df[(student_data_df["grade"] == "11th")] 
grade_12 = student_data_df[(student_data_df["grade"] == "12th")]

#Math scores by schoolper grade
math_score_9th = grade_9.groupby([student_data_df.school_name]).mean()["math_score"]
math_score_10th = grade_10.groupby([student_data_df.school_name]).mean()["math_score"]
math_score_11th = grade_11.groupby([student_data_df.school_name]).mean()["math_score"]
math_score_12th = grade_12.groupby([student_data_df.school_name]).mean()["math_score"]


math_Score_Summary = pd.DataFrame(
    {"9th": math_score_9th,
    "10th": math_score_10th,
    "11th": math_score_11th,
    "12th": math_score_12th
    })

math_Score_Summary = math_Score_Summary[["9th", "10th", "11th", "12th",]]
math_Score_Summary.index.name = None
math_Score_Summary
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#Reading scores by schoolper grade
reading_score_9th = grade_9.groupby([student_data_df.school_name]).mean()["reading_score"]
reading_score_10th = grade_10.groupby([student_data_df.school_name]).mean()["reading_score"]
reading_score_11th = grade_11.groupby([student_data_df.school_name]).mean()["reading_score"]
reading_score_12th = grade_12.groupby([student_data_df.school_name]).mean()["reading_score"]


Reading_Score_Summary = pd.DataFrame(
    {"9th": reading_score_9th,
    "10th": reading_score_10th,
    "11th": reading_score_11th,
    "12th": reading_score_12th
    })

Reading_Score_Summary = Reading_Score_Summary[["9th", "10th", "11th", "12th",]]
Reading_Score_Summary.index.name = None
Reading_Score_Summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Create a table that breaks down school performances based on average Spending Ranges (Per Student). 
# Use 4 reasonable bins to group school spending. Include in the table each of the following:
# Average Math Score
# Average Reading Score
# % Passing Math
# % Passing Reading
# Overall Passing Rate (Average of the above two)

# Create the bins
spending_bins = [0, 585, 615, 645, 675]

# Names for the bins
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

# create dataframe for school spending and create new column
school_summary_data_df["Spending Ranges(Per Student)"] = pd.cut(per_student_budget, bins=spending_bins, labels=group_names)
#school_spending_df.head()

avg_spending_math = school_summary_data_df.groupby(["Spending Ranges(Per Student)"])["Average Math Score"].mean()
avg_spending_reading = school_summary_data_df.groupby(["Spending Ranges(Per Student)"])["Average Reading Score"].mean()
percent_spending_pass_math = school_summary_data_df.groupby(["Spending Ranges(Per Student)"])["% Passing Math"].mean()
percent_spending_pass_reading = school_summary_data_df.groupby(["Spending Ranges(Per Student)"])['% Passing Reading'].mean()
percent_overall_pass = (percent_spending_pass_math + percent_spending_pass_reading)/2

# Creating dataframe
spending_summary = pd.DataFrame ({"Average Math Score": avg_spending_math,
    "Average Reading Score": avg_spending_reading,
    "% Passing Math": percent_spending_pass_math,
    "% Passing Reading": percent_spending_pass_reading,
    "Overall Passing Rate": percent_overall_pass})
spending_summary
```
